```python
#!/usr/bin/env python
import re
import os
import sys
import operator


info_exp=r"ticky: ([\w+]*):? [\w' ]* [\[[0-9#]*\]?]? ?\((.*)\)$"
err_exp=r"ticky: [\w+]*:? ([\w' ]*) [\[[0-9#]*\]?]? ?\((.*)\)$"

def parsing_error(filename,expression):
    '''Takes filename and regular expression as Input and
    returns a dictionary consisting the Error type as key
    and number of occurance as value'''
    error={}
    try:
        with open(filename,'r') as file:
            for line in file.readlines():
                #print(line.strip())
                result=re.search(expression,line.strip())
                if result.group(1) not in error.keys():
                    error[result.group(1)]=1
                else:
                    error[result.group(1)]+=1
    except IOError:
        print("File Read Error.")
    return error






def sort_error(error):
    '''Sort the dictionary in reverse order by value.'''
    return sorted(error.items(), key= operator.itemgetter(1), reverse=True)






def report_error(filename, error, columns):
    '''Generate error report based on the dictionary'''
    try:
        with open(filename, 'w') as file:
            file.write(','.join(columns)+'\n')
            for a,b in error:
                file.write(str(a)+','+str(b)+'\n')
    except IOError:
        print("File Write Error")





def persing_info_error_count(filename,expression):
    '''Return dictionary of Error Count and Info out in reverse order of username'''
    per_user={}
    try:
        with open(filename,'r') as file:
            for line in file.readlines():
                #print(line.strip())
                result=re.search(expression,line.strip())
                if result.group(2) not in per_user.keys():
                    per_user[result.group(2)] = {}
                    per_user[result.group(2)]["INFO"] = 0
                    per_user[result.group(2)]["ERROR"] = 0

                if result.group(1) == "INFO":
                    per_user[result.group(2)]["INFO"] += 1
                elif result.group(1) == "ERROR":
                    per_user[result.group(2)]["ERROR"] += 1
    except IOError:
        print("File Read Error.")
    return per_user





def sort_per_user(per_user):
    '''Sort the dictionary in reverse order by username.'''
    return sorted(per_user.items(), key= operator.itemgetter(0), reverse=True)





def report_per_user(filename, per_user, columns):
    '''Generate error report based on the dictionary'''
    try:
        with open(filename, 'w') as file:
            file.write(','.join(columns)+'\n')
            for value in per_user:
                a,b=value
                file.write(str(a)+','+str(b["INFO"])+','+str(b["ERROR"])+'\n')
    except IOError:
        print("File Write Error")





#Main Function

err_dict=parsing_error(sys.argv[1],err_exp) #Parsing error data from logfile
sort_error(err_dict) #sorting in reverse order by value
report_error(sys.argv[2],sort_error(err_dict),["Error", "Count"]) #generating report

per_user=persing_info_error_count(sys.argv[1],info_exp) #parsing info for Error and Count
report_per_user(sys.argv[3],sort_per_user(per_user),["Username", "INFO", "ERROR"])
```
